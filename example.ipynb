```python
%matplotlib widget
try:
    anime.pause()
except NameError:
    pass

points = np.random.random((5, 2))
velos = np.random.random((5,2))
g = np.array([0, -1])
def update(i):
    global points, velos
    t = 0.05
    points += (velos*t)
    velos += (g*t)
    velos[:,1][points[:,1]<0] *= -1
    velos[:,0][points[:,0]<0] *= -1
    velos[:,0][points[:,0]>1] *= -1
    velos *= 0.995
    plt.cla()
    plt.scatter(*points.T)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.grid(True)
import matplotlib.animation as anim
anime = anim.FuncAnimation(plt.gcf(), update, range(100), interval=50)
plt.show()
```
